# Data dan Preprocessing A2022

```python
pd.set_option('display.max_columns', None)
```

```python
apbd = pd.read_csv('A2022_dataset_buat bedah data.csv', delimiter=';', skiprows=0, low_memory=False)
apbd.head()
```

```python
apbd['namasubfungsi'].fillna('None',inplace=True)
```

```python
apbd['namafungsi'].fillna('None',inplace=True)
```

```python
apbd['namafungsi'].unique()
```

# REALISASI PENDIDIKAN

```python
# filtering the rows where Credit-Rating is Fair
df_3 = apbd[apbd['namafungsi'].str.contains('Pendidikan')]
df_4 = apbd[apbd['namafungsi'].str.contains('PENDIDIKAN')]
```

```python
frames = [df_3, df_4]
result = pd.concat(frames)
```

```python
result.head()
```

```python
data_pend = result[['namapemda','nilaianggaran']]
data_pend
```

```python
data_pend['namapemda'].unique()
```

```python
df3 = data_pend.groupby('namapemda').sum()
df3
```

```python
df3.to_csv('real_pend.csv')
```
